#필요없는 열 삭제

```python
del train_df['PassengerId']
del train_df['Name']
del train_df['Ticket']
del train_df['Cabin']
```

```python
del test_df['PassengerId']
del test_df['Name']
del test_df['Ticket']
del test_df['Cabin']
```

```python
print(train_df)
```

#결측치 처리

```python
train_df.loc[train_df['Sex'] == 'female', 'Age'] = train_df.loc[train_df['Sex'] == 'female', 'Age'].fillna(
    train_df[train_df['Sex'] == 'female']['Age'].mean()
)
```

```python
train_df.loc[train_df['Sex'] == 'male', 'Age'] = train_df.loc[train_df['Sex'] == 'male', 'Age'].fillna(
    train_df[train_df['Sex'] == 'male']['Age'].mean()
)
```

```python
test_df.loc[test_df['Sex'] == 'female', 'Age'] = test_df.loc[test_df['Sex'] == 'female', 'Age'].fillna(
    test_df[test_df['Sex'] == 'female']['Age'].mean()
)
```

```python
test_df.loc[test_df['Sex'] == 'male', 'Age'] = test_df.loc[test_df['Sex'] == 'male', 'Age'].fillna(
    test_df[test_df['Sex'] == 'male']['Age'].mean()
)
```

```python
train_df['Embarked'] = train_df['Embarked'].fillna(train_df['Embarked'].value_counts().index[0])
```

```python
test_df['Embarked'] = test_df['Embarked'].fillna(test_df['Embarked'].value_counts().index[0])
```

범주형 데이터 처리

```python
train_df['Sex'] = train_df['Sex'].replace(['female','male'],[0, 1])
```

```python
test_df['Sex'] = train_df['Sex'].replace(['female','male'],[0, 1])
```

```python
train_df['Pclass'] = train_df['Pclass'].astype('category')
```

```python
train_df['Embarked'] = train_df['Embarked'].astype('category')
```

```python
test_df['Pclass'] = test_df['Pclass'].astype('category')
```

```python
test_df['Embarked'] = test_df['Embarked'].astype('category')
```

카테고리형 데이터 더미변수변환 또는 원핫 인코딩

```python
train_df = pd.get_dummies(train_df)
test_df = pd.get_dummies(test_df)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
x_data=train_df.iloc[:,1:]
print(x_data.head())
```

```python
y_data =train_df['Survived']
print(y_data.head())
```

```python
# 데이터 분할
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
Logistic_model = LogisticRegression()
Logistic_model.fit(x_train, y_train)
```

```python
# 예측
Logistic_pred = Logistic_model.predict(x_test)

# 평가
print(f"Logistic Accuracy: {accuracy_score(y_test, Logistic_pred)}")
print(f"Classification Report:\n{classification_report(y_test, Logistic_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, Logistic_pred)}")
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='linear')
svm_model.fit(x_train, y_train)
```

```python
#예측
svm_pred =svm_model.predict(x_test)
# 평가
print(f"svm Accuracy: {accuracy_score(y_test, svm_pred)}")
print(f"Classification Report:\n{classification_report(y_test, svm_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, svm_pred)}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
Knn_model = KNeighborsClassifier(n_neighbors=5)
Knn_model.fit(x_train, y_train)
```

```python
#예측
Knn_pred =Knn_model.predict(x_test)
# 평가
print(f"KNN Accuracy: {accuracy_score(y_test,Knn_pred)}")
print(f"Classification Report:\n{classification_report(y_test, Knn_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test, Knn_pred)}")
```

```python
from sklearn.naive_bayes import GaussianNB
GaussianNB_model = GaussianNB()
GaussianNB_model.fit(x_train, y_train)
```

```python
#예측
GaussianNB_pred = GaussianNB_model.predict(x_test)
# 평가
print(f"GaussianNB Accuracy: {accuracy_score(y_test,GaussianNB_pred)}")
print(f"Classification Report:\n{classification_report(y_test,GaussianNB_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,GaussianNB_pred)}")
```

```python
from sklearn.tree import DecisionTreeClassifier
DT_model = DecisionTreeClassifier(random_state=42)
DT_model.fit(x_train, y_train)
```

```python
#예측
DT_model_pred= DT_model.predict(x_test)
# 평가
print(f"DT_model_pred Accuracy: {accuracy_score(y_test,DT_model_pred)}")
print(f"Classification Report:\n{classification_report(y_test,DT_model_pred)}")
print(f"Confusion Matrix:\n{confusion_matrix(y_test,DT_model_pred)}")
```
